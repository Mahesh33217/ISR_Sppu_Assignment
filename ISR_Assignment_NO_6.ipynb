# Converting Original img to Red and Yellow Intensity Img

```python
# Read the image
image = cv2.imread("C:\\Users\\Mahesh\\Pictures\\practice\\dashboard_wad\\College Photo 23 Feb 2023.jpg")

# Convert the image from BGR to RGB format (OpenCV uses BGR by default)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Extract red intensity by setting green and blue channels to 0
red_intensity = image_rgb.copy()
red_intensity[:, :, 1] = 0  # Zero out the green channel
red_intensity[:, :, 2] = 0  # Zero out the blue channel

# Extract yellow intensity by combining red and green channels and setting blue to 0
yellow_intensity = image_rgb.copy()
yellow_intensity[:, :, 2] = 0  # Zero out the blue channel

# Plot the original, red intensity, and yellow intensity images
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Original image
axes[0].imshow(image_rgb)
axes[0].set_title('Original Image')
axes[0].axis('off')

# Red intensity image
axes[1].imshow(red_intensity)
axes[1].set_title('Red Intensity')
axes[1].axis('off')

# Yellow intensity image
axes[2].imshow(yellow_intensity)
axes[2].set_title('Yellow Intensity')
axes[2].axis('off')

plt.show()
```

# Original img to Blur Img

```python
# Read the image
image = cv2.imread("C:\\Users\\Mahesh\\Pictures\\practice\\dashboard_wad\\College Photo 23 Feb 2023.jpg")

# Convert the image to RGB (OpenCV loads images in BGR format)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Apply Gaussian Blur
sigma = 20
# OpenCV does not directly accept sigma, but you can compute the kernel size from sigma.
# Kernel size is typically chosen as `6*sigma + 1` for the Gaussian blur.
kernel_size = int(6 * sigma + 1) | 1  # Ensure the kernel size is odd
blurred_image = cv2.GaussianBlur(image_rgb, (kernel_size, kernel_size), sigma)

# Plot the original and blurred images
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Original image
axes[0].imshow(image_rgb)
axes[0].set_title('Original Image')
axes[0].axis('off')

# Blurred image
axes[1].imshow(blurred_image)
axes[1].set_title(f'Blurred Image (sigma={sigma})')
axes[1].axis('off')

plt.show()
```
